# O objetivo desse mini-projeto inicial é estudar tipos de tratamento de datasets com valores vazios. #

```python
import pandas as pd

# criando DataFrame (tabela) de exemplo com dados vazios

data = {
    'Nome': ['Pedro', 'Maria', 'Isabel', 'Vicente', None],
    'Idade': [None, '27', None, '35', '40'],
    'Cidade': ['Recife', None, None, 'Jaboatão','Moreno'],
    'No. de filhos': ['2', None ,'0', None , '1']
}
df = pd.DataFrame(data)
# cria tabela
```

```python
# mostra a tabela criada
display(df)
```

```python
# informações a respeito da tabela

df.info()
```

## Depois de chamar df.info() (responsável por dar insight aos campos vazios da tabela), existem 3 métodos principais de tratar as lacunas. São eles: ##

# 1. Remoção de Linhas e Colunas: dropna() #

Usado para **apenas** remover linhas(0), colunas(1), todos (all) ou qualquer quando não são realmente necessárias (any). Não é muito recomendado por não ser maleável com as informações.

```python

data = {
    'Nome': ['Pedro', 'Maria', 'Isabel', 'Vicente', None],
    'Idade': [None, '27', None, '35', '40'],
    'Cidade': ['Recife', None, None, 'Jaboatão', 'Moreno'],
    'No. de filhos': ['2', None ,'0', None , '1']
}


df = pd.DataFrame(data)
display(df)
df.dropna(axis=0)

#removeu as colunas
```

```python

data = {
    'Nome': ['Pedro', 'Maria', 'Isabel', 'Vicente', None],
    'Idade': [None, '27', None, '35', '40'],
    'Cidade': ['Recife', None, None, 'Jaboatão', 'Moreno'],
    'No. de filhos': ['2', None ,'0', None , '1']
}
df = pd.DataFrame(data)
display(df)
df.dropna(axis=1)

# removeu todas as linhas porque em algumas tinham dados faltando, deixando só os id's
```

# 2. Preenchimento de dados nulos: fillna() #

Preenche os dados faltantes com um valor estabelecido ou pode ser usado com o método ffill que utilizará os valores acima, na mesma coluna. 

```python

data = {
    'Nome': ['Pedro', 'Maria', 'Isabel', 'Vicente', None],
    'Idade': [None, '27', None, '35', '40'],
    'Cidade': ['Recife', None, None, 'Jaboatão', 'Moreno'],
    'No. de filhos': ['2', None ,'0', None , '1']
}
df = pd.DataFrame(data)
display(df)

df.fillna('Desconhecido')

# mudou todos os dados ausentes para "Desconhecido"
```

# 3. Interpolação: interpolate() #

Diferentemente do primeiro e segundo exemplo, a função interpolate não apaga ou altera os dados por um totalmente avulso, mas acha um meio termo entre o valor acima e abaixo da linha com valor vazio. O nome se deve ao conceito matemático da interpolação linear regido pela criação de um novo valor baseado em dois já preexistentes. (interpolação só funciona com dados numéricos).
